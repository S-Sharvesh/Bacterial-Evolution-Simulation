```python
import matplotlib.pyplot as plt

# Sample data for demonstration; replace these with your actual simulation data
time_data = [0, 10, 20, 30, 40, 50]  # Time in seconds
population_data = [100, 95, 90, 85, 80, 75]  # Population of bacteria over time
heat_resistance_data = [1.0, 1.2, 1.4, 1.6, 1.8, 2.0]  # Average heat resistance over time
pressure_resistance_data = [0.8, 1.0, 1.1, 1.3, 1.4, 1.5]  # Average pressure resistance over time

# Plotting the results
plt.figure(figsize=(12, 6))

# Plot 1: Population Growth Over Time
plt.subplot(2, 1, 1)
plt.plot(time_data, population_data, label="Population", color="green", marker="o")
plt.xlabel("Time (seconds)")
plt.ylabel("Population")
plt.title("Population Growth Over Time")
plt.grid(True)
plt.legend()

# Plot 2: Heat and Pressure Resistance Over Time
plt.subplot(2, 1, 2)
plt.plot(time_data, heat_resistance_data, label="Heat Resistance", color="red", marker="o")
plt.plot(time_data, pressure_resistance_data, label="Pressure Resistance", color="blue", marker="o")
plt.xlabel("Time (seconds)")
plt.ylabel("Resistance")
plt.title("Average Heat and Pressure Resistance Over Time")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
